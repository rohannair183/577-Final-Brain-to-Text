## Imports library

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### Input files

```python
cnn_bilstm_trials1 = pd.read_csv('cnn_bilstm_metrics_trial01.csv')
cnn_bilstm_trials2_epoch20 = pd.read_csv('cnn_bilstm_metrics_trial02.csv')
cnn_bilstm_trials2_epoch9 = pd.read_csv('cnn_bilstm_metrics_trial02.csv').iloc[:9,:]
cnn_bilstm_trials2_epoch15 = pd.read_csv('cnn_bilstm_metrics_trial02.csv').iloc[:15,:]
rnn_trials1 = pd.read_csv('rnn_metrics_128_hidden_3_layers_0.001LR.csv')
rnn_trials2 = pd.read_csv('rnn_metrics_128_hidden_3_layers_0.005LR.csv')
```

## CNN BiLSTM vs RNN Best Model

```python
epoch_cnn_bilstm = cnn_bilstm_trials2_epoch9['epoch']
train_loss_cnn_bilstm = cnn_bilstm_trials2_epoch9['train_loss']
val_loss_cnn_bilstm = cnn_bilstm_trials2_epoch9['val_loss']
val_pre_cnn_bilstm = cnn_bilstm_trials2_epoch9['val_per']

epoch_rnn = rnn_trials1['epoch']
train_loss_rnn = rnn_trials1['train_loss']
val_loss_rnn = rnn_trials1['val_loss']
val_pre_rnn = rnn_trials1['val_per']
```

### CNN BiLSTM vs RNN Best Model Metrics Comparison (Train Loss)

```python
# Plot 
plt.figure(figsize=(10, 6))

plt.plot(epoch_cnn_bilstm, train_loss_cnn_bilstm, label='CNN-BiLSTM Train Loss')
plt.plot(epoch_rnn, train_loss_rnn, label='RNN Train Loss')

plt.xlabel("Epoch")
plt.ylabel("Value")
plt.title("CNN BiLSTM vs RNN Best Model Metrics Comparison (Train Loss)")
plt.legend()
plt.grid(True)
plt.show()
```

### CNN BiLSTM vs RNN Best Model Metrics Comparison (Val Loss)

```python
# Plot 
plt.figure(figsize=(10, 6))

plt.plot(epoch_cnn_bilstm, val_loss_cnn_bilstm, label='CNN-BiLSTM Val Loss')
plt.plot(epoch_rnn, val_loss_rnn, label='RNN Val Loss')

plt.xlabel("Epoch")
plt.ylabel("Value")
plt.title("CNN BiLSTM vs RNN Best Model Metrics Comparison (Val Loss)")
plt.legend()
plt.grid(True)
plt.show()
```

### CNN BiLSTM vs RNN Best Model Metrics Comparison (Val Per)

```python
# Plot 
plt.figure(figsize=(10, 6))

plt.plot(epoch_cnn_bilstm, val_loss_cnn_bilstm, label='CNN-BiLSTM Val Per')
plt.plot(epoch_rnn, val_loss_rnn, label='RNN Val Per')

plt.xlabel("Epoch")
plt.ylabel("Value")
plt.title("CNN BiLSTM vs RNN Best Model Metrics Comparison (Val Per)")
plt.legend()
plt.grid(True)
plt.show()
```
