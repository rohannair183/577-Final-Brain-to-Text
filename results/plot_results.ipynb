## Imports library

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### Input files

```python
cnn_bilstm_trials1 = pd.read_csv('cnn_bilstm_metrics_trial01.csv')
cnn_bilstm_trials2_epoch20 = pd.read_csv('cnn_bilstm_metrics_trial02.csv')
cnn_bilstm_trials2_epoch9 = pd.read_csv('cnn_bilstm_metrics_trial02.csv').iloc[:9,:]
cnn_bilstm_trials2_epoch15 = pd.read_csv('cnn_bilstm_metrics_trial02.csv').iloc[:15,:]
rnn_trials1 = pd.read_csv('rnn_metrics_128_hidden_3_layers_0.001LR.csv')
rnn_trials2 = pd.read_csv('rnn_metrics_128_hidden_3_layers_0.005LR.csv')
tcn_trials1 = pd.read_csv('tcn_metrics_trial01.csv')
tcn_trials1_epoch15 = pd.read_csv('tcn_metrics_trial01.csv').iloc[:15,:]
```

## CNN BiLSTM vs RNN vs TCN Best Model

```python
epoch_cnn_bilstm = cnn_bilstm_trials2_epoch15['epoch']
train_loss_cnn_bilstm = cnn_bilstm_trials2_epoch15['train_loss']
val_loss_cnn_bilstm = cnn_bilstm_trials2_epoch15['val_loss']
val_pre_cnn_bilstm = cnn_bilstm_trials2_epoch15['val_per']

epoch_rnn = rnn_trials1['epoch']
train_loss_rnn = rnn_trials1['train_loss']
val_loss_rnn = rnn_trials1['val_loss']
val_pre_rnn = rnn_trials1['val_per']

epoch_tcn = tcn_trials1_epoch15['epoch']
train_loss_tcn = tcn_trials1_epoch15['train_loss']
val_loss_tcn = tcn_trials1_epoch15['val_loss']
val_pre_tcn = tcn_trials1_epoch15['val_per']
```

### CNN BiLSTM vs RNN vs TCN Best Model Metrics Comparison (Train Loss)

```python
# Plot 
plt.figure(figsize=(10, 6))

plt.plot(epoch_cnn_bilstm, train_loss_cnn_bilstm, label='CNN-BiLSTM Train Loss')
plt.plot(epoch_rnn, train_loss_rnn, label='RNN Train Loss')
plt.plot(epoch_tcn, train_loss_tcn, label='TCN Train Loss')

plt.xlabel("Epoch")
plt.ylabel("Train Loss")
plt.yticks(np.arange(1, 8, 1))
plt.title("CNN BiLSTM vs RNN vs TCN Best Model Metrics Comparison (Train Loss)")
plt.legend()
plt.grid(True)
plt.show()
```

### CNN BiLSTM vs RNN vs TCN Best Model Metrics Comparison (Val Loss)

```python
# Plot 
plt.figure(figsize=(10, 6))

plt.plot(epoch_cnn_bilstm, val_loss_cnn_bilstm, label='CNN-BiLSTM Val Loss')
plt.plot(epoch_rnn, val_loss_rnn, label='RNN Val Loss')
plt.plot(epoch_tcn, val_loss_tcn, label='TCN Val Loss')

plt.xlabel("Epoch")
plt.ylabel("Val Loss")
plt.yticks(np.arange(1.5, 3.3, 0.2))
plt.title("CNN BiLSTM vs RNN vs TCN Best Model Metrics Comparison (Val Loss)")
plt.legend()
plt.grid(True)
plt.show()
```

### CNN BiLSTM vs RNN vs TCN Best Model Metrics Comparison (Val Per)

```python
# Plot 
plt.figure(figsize=(10, 6))

plt.plot(epoch_cnn_bilstm, val_pre_cnn_bilstm, label='CNN-BiLSTM Val PER')
plt.plot(epoch_rnn, val_pre_rnn, label='RNN Val PER')
plt.plot(epoch_tcn, val_pre_tcn, label='TCN Val PER')

plt.xlabel("Epoch")
plt.ylabel("Val PER")
plt.yticks(np.arange(0.3, 0.8, 0.05))
plt.title("CNN BiLSTM vs RNN vs TCN Best Model Metrics Comparison (Val PER)")
plt.legend()
plt.grid(True)
plt.show()
```
